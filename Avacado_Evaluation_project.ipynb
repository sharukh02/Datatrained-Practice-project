## DataCleaning

```python
plt.figure(figsize=(25,30))
plotnumber = 1
for i in df:
    if plotnumber <18:
        ax = plt.subplot(6,3,plotnumber)
        sns.boxplot(df[i])
        plt.xlabel(i,fontsize=12)
        plotnumber+=1
plt.show()
```

Total Volume, 4046, 4225, 4770, Total Bags, Small Bags, Large Bags, XLarge Bags these columns has Outliers.

```python
features = ['Total Volume', '4046', '4225', '4770','Total Bags', 'Small Bags', 'Large Bags', 'XLarge Bags']
```

```python
from scipy.stats import zscore
```

```python
z_score = zscore(df[features])
abs_z_score = np.abs(z_score)
```

```python
filter = ((abs_z_score <3) & (abs_z_score > -3)).all(axis=1)
```

```python
new_df = df[filter]
```

```python
print("The Data loss is :",((df.shape[0]-new_df.shape[0])/df.shape[0])*100)
```

```python
datacleaned = new_df[new_df["Total Volume"] < new_df["Total Volume"].quantile(0.92)]
datacleaned = new_df[new_df["4046"] < new_df["4046"].quantile(0.92)]
datacleaned = new_df[new_df["4225"] < new_df["4225"].quantile(0.92)]
datacleaned = new_df[new_df["4770"] < new_df["4770"].quantile(0.92)]
datacleaned = new_df[new_df["Total Bags"] < new_df["Total Bags"].quantile(0.92)]
datacleaned = new_df[new_df["Small Bags"]< new_df["Small Bags"].quantile(0.92)]
datacleaned = new_df[new_df["Large Bags"]< new_df["Large Bags"].quantile(0.92)]
datacleaned = new_df[new_df["XLarge Bags"]< new_df["XLarge Bags"].quantile(0.90)]
```

```python
x = df.drop(columns = ["AveragePrice","Unnamed: 0"],axis=1)
y= df["AveragePrice"]
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.ensemble import GradientBoostingRegressor,BaggingRegressor,AdaBoostRegressor,RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler,PowerTransformer
from sklearn.model_selection import train_test_split,cross_val_score,KFold,LeaveOneOut
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.decomposition import PCA
```

```python
scale = StandardScaler()
x_scaled = scale.fit_transform(x)
```

```python
max_accu = 0
max_RS = 0
for i in range(1,300):
    x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30,random_state=i)
    mod = XGBRegressor()
    mod.fit(x_train,y_train)
    y_pred = mod.predict(x_test)
    accu_score = r2_score(y_test,y_pred)
    if accu_score > max_accu:
        max_Rs = i
        max_accu = accu_score
print("Random state :",max_RS,"Best accuracy :",max_accu)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.30,random_state=0)
```

```python
kf = KFold(n_splits = 15,shuffle=False,random_state=0)
```

## LinearRegression

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
r = r2_score(y_test,y_pred)
c = cross_val_score(lr,x_scaled,y,cv=kf).mean()
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
mae = mean_absolute_error(y_test,y_pred)
```

```python
print(" R2 score :",r,"\n","="*80,"\n Cross_validation Score :",c,"\n","="*80,"\n Mean Squared Error :",mse,"\n","="*80,
     "\n Mean Absolute Error :",mae,"\n","="*80,"\n Root Mean Squared Error",rmse)
```

```python
r2 = []
cv = []
ms = []
r2.append(r)
cv.append(c)
ms.append(mse)
```
