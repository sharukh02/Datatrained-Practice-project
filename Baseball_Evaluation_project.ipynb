## DataCleaning

```python
plt.figure(figsize=(16,28),facecolor='white')
plotnumber = 1
for i in df:
    if plotnumber < 19:
        ax= plt.subplot(9,2,plotnumber)
        sns.boxplot(df[i])
        plotnumber+=1
plt.show()
```

R, ERA, SHO, SV and E has Outliers.

```python
features=["R","ERA","SHO","SV","E"]
```

```python
# Applying z Score method to remove Outliers
from scipy.stats import zscore
```

```python
z_score=zscore(df[features])
abs_z_score=np.abs(z_score)
filtered_entries = ((abs_z_score < 3 ) & (abs_z_score > -3 )).all(axis=1)
new_df = df[filtered_entries]
new_df.head()
```

```python
plt.figure(figsize=(16,28),facecolor='white')
plotnumber = 1
for i in new_df:
    if plotnumber < 19:
        ax= plt.subplot(9,2,plotnumber)
        sns.boxplot(new_df[i])
        plotnumber+=1
plt.show()
```

```python
# Applying quantile method to take care further Outliers
datacleaned = new_df[new_df["SHO"] < new_df["SHO"].quantile(0.99)]
datacleaned = new_df[new_df["H"] < new_df["H"].quantile(0.95)]
datacleaned = new_df[new_df["E"] < new_df["E"].quantile(0.97)]
datacleaned = new_df[new_df["SV"] < new_df["SV"].quantile(0.96)]
```

```python
datacleaned.shape
```

```python
plt.figure(figsize=(16,28),facecolor='white')
plotnumber = 1
for i in datacleaned:
    if plotnumber < 19:
        ax= plt.subplot(9,2,plotnumber)
        sns.boxplot(datacleaned[i])
        plt.xlabel(i,fontsize=10)
        plotnumber+=1
plt.show()
```

```python
# Making a copy of a Dataframe
df1 = pd.DataFrame(new_df)
```

```python
#Trying Interquartile method
q1 = df1.quantile(0.25)
q3 = df1.quantile(0.75)
iqr = q3-q1
```

```python
for column in df1[features]:
    max=(q3[column]+(1.5*iqr[column]))
    index=np.where(df1[column]>max)
    df1.drop(df1.index[index],inplace=True,axis=0)
df1.shape
```

#### I have select Best Using KBest Because here we have very less Dataset. And Machine learning models are datahungry so if we pass all the feature for limited observation it may be overfit the model.

Therefore i have select the Best feature to avoid overfitting but still there is chances that model can give high variance because there is not sufficient data for trainings and testing

```python
from sklearn.preprocessing import StandardScaler,PowerTransformer
from sklearn.model_selection import cross_val_score,train_test_split,LeaveOneOut,GridSearchCV,RandomizedSearchCV
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression,Lasso,Ridge,LassoCV,RidgeCV
from sklearn.ensemble import GradientBoostingRegressor,AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.decomposition import PCA
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
x= datacleaned[feature]
y= datacleaned["W"]
```

```python
scale = StandardScaler()
x_scaled = scale.fit_transform(x)
```

```python
max_accu = 0
max_RS = 0
for i in range(1,300):
    x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30,random_state=i)
    mod = Lasso()
    mod.fit(x_train,y_train)
    y_pred = mod.predict(x_test)
    accu_score = r2_score(y_test,y_pred)
    if accu_score > max_accu:
        max_Rs = i
        max_accu = accu_score
print("Random state :",max_RS,"Best accuracy :",max_accu)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.30,random_state=0)
```
