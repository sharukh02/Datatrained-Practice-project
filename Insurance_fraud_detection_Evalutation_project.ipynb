```python
### Applying z score
from scipy.stats import zscore
zscr = zscore(df[["age","policy_annual_premium","umbrella_limit","property_claim","total_claim_amount"]])
```

```python
abs_zscr = np.abs(zscr)
filter = ((abs_zscr < 3) & (abs_zscr > -3)).all(axis=1)
```

```python
new_df = df[filter]
```

```python
new_df.shape
```

```python
# Removing Outliers from some columns
q3 = new_df.quantile(0.75)
q1 = new_df.quantile(0.25)
iqr = q3-q1
max = q3["age"] + (1.5*iqr["age"])
index = np.where(new_df["age"] > max)
new_df.drop(new_df.index[index],inplace=True)
```

```python
# Removing Outliers from policy annual premium
max = q3["policy_annual_premium"] + (1.5*iqr["policy_annual_premium"])
index = np.where(new_df["policy_annual_premium"] > max)
new_df.drop(new_df.index[index],inplace=True)

min = q1["policy_annual_premium"] - (1.5*iqr["policy_annual_premium"])
index = np.where(new_df["policy_annual_premium"] < min)
new_df.drop(new_df.index[index],inplace=True)
```

```python
# Removing Outliers from policy annual premium
max = q3["property_claim"] + (1.5*iqr["property_claim"])
index = np.where(new_df["property_claim"] > max)
new_df.drop(new_df.index[index],inplace=True)
```

```python
# Removing Outliers from total_claim_amount
max = q3["total_claim_amount"] + (1.5*iqr["total_claim_amount"])
index = np.where(new_df["total_claim_amount"] > max)
new_df.drop(new_df.index[index],inplace=True)
```

```python
min = q1["umbrella_limit"] - (1.5*iqr["umbrella_limit"])
index = np.where(new_df["umbrella_limit"] < min)
new_df.drop(new_df.index[index],inplace=True)
```
