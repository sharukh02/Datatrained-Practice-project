```python
features = ["Capital_loss","Capital_gain","Fnlwgt"]
```

```python
# Creating a copy of dataframe
new_df = pd.DataFrame(df)
```

```python
q1 = new_df.quantile(0.25)
q3 = new_df.quantile(0.75)
iqr = q3 - q1
```

```python
max = q3["Capital_loss"]+(1.5+iqr["Capital_loss"])
index = np.where(new_df["Capital_loss"] > max)
new_df.drop(new_df.index[index],inplace=True)
```

```python
max = q3["Capital_gain"]+(1.5+iqr["Capital_gain"])
index = np.where(new_df["Capital_gain"] > max)
new_df.drop(new_df.index[index],inplace=True)
```

```python
## Using smote to balances the Target columns
from sklearn.utils import resample
zero = new_df[new_df["Income"]==0]# Less than 50k
one = new_df[new_df["Income"]==1]# Greater than 50k
```

```python
up_sampled = resample(one,replace=True,n_samples=len(zero),random_state=0)
```

```python
new_df=pd.concat([zero,up_sampled])
```

```python
new_df["Income"].value_counts()
```

```python
sns.countplot(new_df["Income"],data=new_df)
```

Class imbalancement has been taken cared

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split,cross_val_score,KFold
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier, BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.decomposition import PCA
from sklearn.svm import SVC
```

```python
x = new_df.drop(columns=["Income"],axis=1)
y = new_df["Income"]
```

```python
scale = StandardScaler()
x_scaled = scale.fit_transform(x)
```

```python
# Fixing the random state in cross validation to stop fluctuation in cv score every time we run
kf = KFold(n_splits=5,random_state=114)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.30,random_state=114)
```

```python
diff= []
cross = []
accu = []
```

### LogisticRegression

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
a = accuracy_score(y_test,y_pred)
```

```python
c = cross_val_score(lr,x_scaled,y,cv=kf).mean()
```

```python
accu.append(a)
cross.append(c)
diff.append(a-c)
```

```python
print("Accuracy score is :",a,"\n","=-"*60,"\nCross_validation_score is :",c,"\n","=-"*60,"\n Classification Report :\n",
      classification_report(y_test,y_pred),"\n","=-"*60,"\nConfusion Matrix :\n",confusion_matrix(y_test,y_pred))
```
