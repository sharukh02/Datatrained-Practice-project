```python
from sklearn.utils import resample
not_survived = new_df1[new_df1["Survived"]==0]
survived = new_df1[new_df1["Survived"]==1]
```

```python
up_sampled = resample(survived,replace=True,n_samples=len(not_survived),random_state=0)
```

```python
new_df=pd.concat([not_survived,up_sampled])
new_df
```

Dataset is balanced. It's not required but still i am balancing so that model have all kind of data

```python
x = new_df.drop(["Survived","PassengerId","Ticket_freq"],axis=1)
y= new_df["Survived"]
```

PassengerId here is Just like a index no.(Serial no.) So i am dropping in from the dataset

```python
from sklearn.preprocessing import PowerTransformer
```

```python
pt = PowerTransformer()
x[["Fare"]] = pt.fit_transform(x[["Fare"]])
```

```python
x.skew()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
x_scaled = scale.fit_transform(x)
```

```python
# Importing models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,BaggingClassifier,GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV,RandomizedSearchCV
from sklearn.metrics import accuracy_score,classification_report,roc_auc_score,confusion_matrix
```

```python
max_accu = 0
max_RS = 0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30,stratify=y,random_state=i)
    mod = GradientBoostingClassifier()
    mod.fit(x_train,y_train)
    y_pred = mod.predict(x_test)
    accu_score = accuracy_score(y_test,y_pred)
    if accu_score > max_accu:
        max_Rs = i
        cv = cross_val_score(mod,x_scaled,y,cv=5).mean()
        max_accu = accu_score
print("Random state :",max_RS,"Best accuracy :",max_accu,"Cross validation score :",cv)
```
